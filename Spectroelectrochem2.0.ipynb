```python
#561.980835 seemed to be the peak wl
norm255=(Abs255[561.980835]-Abs255[544.9674072])
norm270=(Abs270[561.980835]-Abs270[544.9674072])
norm246=(Abs246[561.980835]-Abs246[544.9674072])
norm190=(Abs190[561.980835]-Abs190[544.9674072])
norm22=(Abs22[561.980835]-Abs22[544.9674072])
norm_31=(Abs_31[561.980835]-Abs_31[544.9674072])
norm_174=(Abs_174[561.980835]-Abs_174[544.9674072])
norm_192=(Abs_192[561.980835]-Abs_192[544.9674072])
norm_214=(Abs_214[561.980835]-Abs_214[544.9674072])
norm_245=(Abs_245[561.980835]-Abs_245[544.9674072])
norm_261=(Abs_261[561.980835]-Abs_261[544.9674072])
norm_275=(Abs_275[561.980835]-Abs_275[544.9674072])
norm_309=(Abs_309[561.980835]-Abs_309[544.9674072])
norm_375=(Abs_375[561.980835]-Abs_375[544.9674072])
norm_469=(Abs_469[561.980835]-Abs_469[544.9674072])
norm_307=(Abs_307[561.980835]-Abs_307[544.9674072])
norm_271=(Abs_271[561.980835]-Abs_271[544.9674072])
norm_241=(Abs_241[561.980835]-Abs_241[544.9674072])
norm_212=(Abs_212[561.980835]-Abs_212[544.9674072])
norm_177=(Abs_177[561.980835]-Abs_177[544.9674072])
norm_66=(Abs_66[561.980835]-Abs_66[544.9674072])
norm_16=(Abs_16[561.980835]-Abs_16[544.9674072])
```

```python
y=[norm255,norm270,norm246,norm190,norm22,norm_16,norm_31,norm_66,norm_174,norm_177,norm_192,norm_212,norm_214,norm_241,norm_245,norm_261,norm_271,norm_275,norm_307,norm_309,norm_375,norm_469]

x=[255,270,246,190,22,-16,-31,-66,-174,-177,-192,-212,-214,-241,-245,-261,-271,-275,-307,-309,-375,-469]
scatter(x,y)

def nernst(x, Em, z):
    return (0.7/(10**((x-Em)/(8.314*273.15/96.5))+1))+z
guess=[-280, -1]
popt, pcov=curve_fit(nernst,x,y,guess)
print(popt)
```

```python
plot(x,nernst(x,*popt), label='fitted')
scatter(x,y)
```
