```python
def nernst(x, Em):
    return 100/(10.0**((x-Em)/(8.314*273.15/96.5))+1)#this is where I'm stuck, the equation seems to work mathematically, but it's not fitting well at all
guess=[-280]
popt, pcov=curve_fit(nernst,x,y,guess)
print(popt)
```

```python
plot(x,nernst(x,*popt), label='fitted')
scatter(x,y)
```
